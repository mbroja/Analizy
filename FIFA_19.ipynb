## Analiza danych zaczerpniętych z gry FIFA 19.

Dane zawierają informację na temat poszczególnych piłkarzy, ich wieku, wartości na rynku, punktów umiejętności w skali o 0 do 100 itd.

Importuję potrzebne biblioteki oraz plik w formacie csv.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
fifa = pd.read_csv('data.csv')
```

```python
fifa.head()
```

```python
fifa.info()
```

#### Wstępna analiza zastanych danych.
Dzięki komendzie head() oraz info() widzimy, że istnieje ponad 18 tysięcy rekordów oraz 89 kolumn. Wiele z tych danych nie będzie nam potrzebnych w poźniejszej analizie, dlatego pozbędziemy się większości z typem object.

```python
fifa_2 = fifa[['Name', 'Age', 'Nationality', 'Overall', 'Potential', 'Club', 'Value', 'Wage', 'Preferred Foot', 'International Reputation', 'Skill Moves', 'Work Rate', 'Body Type', 'Position', 'Jersey Number', 'Height', 'Weight']]
fifa_3 = fifa[fifa.columns[54:88]]
fifa_4 = fifa_2.join(fifa_3)
fifa_4.head()
```

#### Czyszczenie danych
Po usunięciu niepotrzebnych kolumn, do dalszej analizy będą potrzebne dane liczbowe w "czystszej" formie, niż ta przedstawiona powyżej. Zmienne takie jak 'Value' oraz ' Wage' zostaną pozbawione waluty występującej przed wartością oraz w zależności od literki 'K' lub 'M', zostaną pomnożone odpowiednio przez tysiąc lub milion. 
Dodatkowo sprawdzamy, czy są jakieś brakujące wartości.

```python
fifa_4.isnull().values.any()
```

```python
plt.figure(figsize = (10,10))

sns.heatmap(fifa_4.isnull())
```

```python
nan_rows = fifa_4[fifa_4['Preferred Foot'].isnull()]
print(nan_rows)
```

Jest około 50 piłkarzy z prakującymi danymi, jako że brakuje im większości kolum i nie jest to duża liczba, po porstu je usunę.

```python
fifa_4 = fifa_4.dropna()
```

Funkcja dla kolumny: 'Value', ' Wage':

```python
def f(x):
    if "M" in list(x):
        return float(str(x)[1:-1]) * 1000000
    elif "K" in list(x):
        return float(str(x)[1:-1]) * 1000
    else:
        return float(str(x)[1:])
```

```python
fifa_4['Value'] = fifa_4['Value'].map(f)
fifa_4['Wage'] = fifa_4['Wage'].map(f)
fifa_4.head()
```

Funkcja dla zmiennej 'Height':

```python
def to_cm(x):
    return str((float(x[:1]) * 12 + float(x[2:])) * 2.54)
```

```python
fifa_4['Height'] = fifa_4['Height'].map(to_cm)
```

Funkcja dla zmienniej Weight:

```python
def to_kg(x):
    return float(str(x)[:-3]) * 0.45359237
```

```python
fifa_4['Weight'] = fifa_4['Weight'].map(to_kg)
```

#### Proste statystyki
Ogólne, piłkarze z najlepszym potencjałem, z największą wartością

```python
fifa_4['Wage'] = fifa_4['Wage'].astype('int64')
fifa_4['Value'] = fifa_4['Value'].astype('int64')
fifa_4['Weight'] = fifa_4['Weight'].astype('int64')
```

## Model dla regresji liniowej

```python
x = fifa_4[['Potential', 'Wage', 'Overall', 'International Reputation', 'Skill Moves']]
y = fifa_4['Value']
x.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.4)
```

```python
x_train.info()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
```

```python
lm.fit(x_train, y_train)
```

```python
coeff_df = pd.DataFrame(lm.coef_, x.columns, columns = ['Coefficient'])
```

#### Wartość zawodnika rośnie wraz ze wzrostem każdego z poszczególnych czynników, co ma sens zwłaszcza w przypadku zmiennej Overall

```python
print(coeff_df)
```

```python
predictions = lm.predict(x_test)
```

```python
plt.figure(figsize = (10,10))

plt.title("Predykcja")
plt.scatter(y_test, predictions)
```

```python
from sklearn import metrics
```

```python
print('MAE:', metrics.mean_absolute_error(y_test, predictions))
print('MSE:', metrics.mean_squared_error(y_test, predictions))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, predictions)))
```
